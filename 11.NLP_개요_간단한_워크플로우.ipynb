#### 분절화(형태소단위)

- a. 원본 문장
```
설정이 재밌고 새로운 에피소드 내에서 메인 스토리도 차차 나오는게 재밌음
```

- b.형태보 분절하여 다시 공백 단위 문자
```
설정 이 재밌고 새로운 에피소드 내 에서 메인 스토리 도 차차 나오는게 재밌음
```

- 한국어 특성에 맞게 a문장을 b문장으로 조정 -> keras의 토크나이저 도구(공백 단위 분절)를 사용 -> 나머지 절차 진행

````python
train_df_mini = pd.DataFrame()
test_df_mini = pd.DataFrame()
````

````python
%%time
# 5. 기준 문장 => 분절화 => 공백단위 조립 => 원문 완성
# 실습 훈련용 15000개 문장, 정답 (앞에서부터 자름)
train_df_mini['sentence'] = train_df.sentence[:15000].apply(
                        lambda doc:' '.join( tokenizer.morphs( doc ) ) )
train_df_mini['label']    = train_df.label[:15000]

# 실습 테스트용 5000개 문장, 정답 (앞에서부터 자름)
test_df_mini['sentence'] = test_df.sentence[:5000].apply(
                        lambda doc:' '.join( tokenizer.morphs( doc ) ) )
test_df_mini['label']    = test_df.label[:5000]
````

- 자연어 처리쪽 백본, LLM쪽은 공백 기반 분절로 사용
    - 대량의 데이터로 학습

````python
train_df_mini.head(2)
````

#### 분절화(공백단위 진행)

````python
from tensorflow.keras.preprocessing.text import Tokenizer

# 6. 공백기반 토크나이저(분절, 기타 다양한기능( 사전화, 패딩,.)  획득
nlp_tokenizer = Tokenizer()
````

````python
%%time
# 7. 문장 => 분절 => 토큰분해진행 => 사전화 구축
nlp_tokenizer.fit_on_texts( train_df_mini.sentence )
````

#### 사전화

````python
# 8. 사전화 결과 확인
# 사전화 과정에 사용되느 총 문장의 수
print( nlp_tokenizer.document_count )
#print( nlp_tokenizer.num_words )
# { 토큰번호:토큰, ... }, {1: '이', 2: '영화', ...
print( len(nlp_tokenizer.index_word) )  # 27448 개의 사전 구축
````

#### 백터화

````python
# 9. 문장 => 수치로표현 => [ 4, 3, 5, 6, ...]:백터
doc_vectors = nlp_tokenizer.texts_to_sequences( train_df_mini.sentence )
len(doc_vectors), type(doc_vectors)
````

#### 패딩

- 최대 길이로 지정
- API 사용
- 세팅값
    - 통상 0
        - 문장 -> 백터화 -> 패딩 수행시
    - 스페셜토큰 <pad> 방식 존재 -> 수치화 진행
        - 문장 -> 패딩 -> 백터화

````python
# 10. 백터 문장들중 가장 큰 길이 획득
# 실습 1분
max_len = max( [ len(vector) for vector in doc_vectors ] )

# 72 => 모든 문장은 1+72 패딩 진행
# +1 => 케라스를 이용할 때 맨 앞자리는 무조건 0으로 두고 채운다는 의미
max_len
````

````python
from tensorflow.keras.preprocessing.sequence import pad_sequences

# 11. 패딩 진행
# padding='pre' <- 앞쪽으로 패딩
# truncating='pre' <- 지정한 최대길이보다 재료 백터가 넘치면 앞에서 자른다
# 언어별 관습에 따라 절단 위치를 조정할수 있다
padded_x = pad_sequences( doc_vectors, 1+max_len )
padded_x.shape, type(padded_x)
````

### 인공신경망구축및학습

- 기본형 순환신경망 사용
    - LSTM 기반 진행
    - 간단하게 구성 - 성능 고려 X
    - tensorflow/keras 로 구현

- 인공신경망 구축및 학습 파트와 다른 부분은 분리하는게 유리

##### 인공신경망 구축

````python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Embedding, Dense, LSTM
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
````

````python
# 사전에 등록된 토큰의 총개수 + 패딩용 1개(보정용)
token_count = len( nlp_tokenizer.index_word) + 1
token_count
````

````python
# 12. 모델(순환신경망 형태) 구축
model = Sequential()

model.add( Embedding( token_count, 1+max_len) )
model.add( LSTM(128) ) # 128 설정
model.add( Dense(1, activation='sigmoid') ) # 이진분류, 긍정1,부정0

model.summary()
````

##### 모델 컴파일 구성

````python
# 13. 컴파일 옵션 구성
model.compile(optimizer='adam',loss='binary_crossentropy',
              metrics=['accuracy'] )
````

##### 모델 학습

- 자연어 학습 시간 CPU대비 GPU의 속도가 급격하게 차이가 나지는 않음

````python
import tensorflow as tf
# 14. 학습
# 실습: GPU 설정, 학습 파라미터 설정,
#      세대학습 10, 배치사이즈 64, 콜백(조기학습종료, 모델체포인트),
#      검증용 데이터 비율 0.2 지정
es = EarlyStopping(patience=4, verbose=1, mode='min')
mc = ModelCheckpoint('best_nlp.h5', mode='max', verbose=1,
                      save_best_only=True)
with tf.device('/device:GPU:0'):
    history = model.fit( padded_x,
                        train_df_mini.label,
                        batch_size = 64,
                        epochs     = 10,
                        callbacks  = [es, mc],
                        validation_split = 0.2
                        )
````

````python
import gradio

# 2. 메인 기능 함수 구성
def sentence_review_pos_neg_clf( msg, history ):
    '''
        - msg : 사용자 입력 문장
        - history : 채팅 히스토리
    '''
    # 토크나이저의 역활
    test_text = test_df_mini.sentence[10]
    msg     = tokenizer.morphs( msg )
    vec     = nlp_tokenizer.texts_to_sequences( [ msg ] )
    pad     = pad_sequences( vec, 1+72) # 1+max_len)
    # 모델의 역활
    pred    = model_loaded.predict( pad )
    # 결과값 리턴
    if pred[0][0] >= 0.5:
        return f'긍정, {pred[0][0]}'
    return f'부정, {pred[0][0]}'


# 3. 데모 구성
demo = gradio.ChatInterface( sentence_review_pos_neg_clf )

# 4. 데모 런치
demo.launch()

# 훈련에 사용된 토큰이 아닌 표현 => 사전에 등록 X => 판단과정에서 부정확성을 유발 => LLM
````
